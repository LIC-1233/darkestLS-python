```python
from lark import Lark

grammar = r'''
start: comment*
comment: COMMENT
COMMENT: /\/[^\/]+\n?/

NEWLINE2:  /\s*/ /\r?/ /\n/
TEST: "1"
%ignore NEWLINE2
%ignore TEST
'''

text = r'''/123    /1123'''

la = Lark(grammar,parser='lalr',propagate_positions=True)
pt = la.parse(text) 
print(pt)
print( pt.pretty() )
```
